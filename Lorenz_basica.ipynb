```python
import numpy as np
import matplotlib.pyplot as plt

# Definir os pontos da curva de Lorenz teórica (desigualdade)
x = np.linspace(0, 1, 100)
y = x**2  # Curva de Lorenz teórica para um exemplo de desigualdade

# Calcular o índice de Gini como a área entre a diagonal e a curva de Lorenz
AUC_Lorenz = np.trapz(y, x)  # Área sob a curva de Lorenz
AUC_total = 0.5  # Área sob a linha de igualdade (triângulo base 1 e altura 1)
gini = (AUC_total - AUC_Lorenz) / AUC_total  # Índice de Gini

# Criar a figura e o gráfico
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x, y, label="Curva de Lorenz", color="blue", linewidth=2)
ax.plot(x, x, label="Igualdad perfecta", color="black", linestyle="dashed")

# Adicionar a curva da desigualdade extrema (ideal para concentração máxima)
ax.plot([0, 1, 1], [0, 0, 1], label="Desigualdad extrema", color="red", linewidth=2)

# Preencher a área de Gini
ax.fill_between(x, y, x, color="lightblue", alpha=0.5, label="Área de Gini")

# Adicionar os pontos "A" e "B"
ax.scatter(0.2, 0.2, color="brown", s=100, label="Punto A")
ax.text(0.2, 0.2, " A", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color="brown")

ax.scatter(0.8, 0.8, color="green", s=100, label="Punto B")
ax.text(0.8, 0.8, " B", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color="green")

ax.scatter(0.2, 0.04, color="blue", s=100, label="Punto C")
ax.text(0.2, 0.04, " C", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color="blue")

ax.scatter(0.8, 0.64, color="orange", s=100, label="Punto D")
ax.text(0.8, 0.64, " D", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color="orange")

ax.scatter(1, 1, color="red", s=100, label="Punto M")
ax.text(1, 1, " M", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color="red")

# Etiquetas e título
ax.set_xlabel("Porcentaje acumulado de población")
ax.set_ylabel("Porcentaje acumulado de ingresos")
ax.set_title(f"Curva de Lorenz y índice de Gini: {gini:.2f}")
ax.legend()
ax.grid(True)

# Mostrar o gráfico
plt.show()
```
